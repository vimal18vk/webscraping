```python
import requests
```

```python
from bs4 import BeautifulSoup

r = requests.get("https://www.indiatoday.in/")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
print(soup.prettify())
```

```python
all=soup.find_all("ul",{"class":"itg-listing"})
all
```

```python
len(all)
```

```python
new = all[0].find_all("li")
new
```

```python
new[0].find_all("a")
```

```python
for item in new:   #to display all price value in the website
    print(item.find_all("a")[0].text)
```

```python
l=[]
for item in new:   #to display all price value in the website
    d={}
    d["top_news"]=item.find_all("a")[0].text
    
    l.append(d)
```

```python
l
```

```python
len(l)
```

```python
import pandas
df=pandas.DataFrame(l)
```

```python
df
```

```python
df.to_csv("topstories_3_03_2022.csv")
```
